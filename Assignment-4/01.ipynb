```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Import the football dataset from Kaggle
df = pd.read_csv('./Downloads/football_data.csv')

df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
# Plot a bar graph of the number of goals scored by each team
goals_by_team = df.groupby('player_club_id')['goals'].sum().reset_index()
goals_by_team = goals_by_team.sort_values(by='goals', ascending=False)
```

```python
plt.figure(figsize=(8, 4))
plt.bar(goals_by_team['player_club_id'], goals_by_team['goals'])
plt.xticks(rotation=45)
plt.xlabel('Teams ID')
plt.ylabel('Total Goals')
plt.title('Total Goals Scored by Each Team')
plt.show()
```

```python
# Plot a scatter plot of the number of shots taken and the number of goals scored
plt.scatter(x=df['minutes_played'], y=df['goals'])
plt.xlabel('Number of minutes played')
plt.ylabel('Number of goals scored')
plt.title('Number of minutes played vs. number of goals scored')
plt.show()
```

```python
# Plot a box plot of the distribution of the number of passes made by each team
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['minutes_played'])
plt.xlabel('Players')
plt.ylabel('Minutes played')
plt.title('Distribution of the number of minutes played by each player')
plt.show()
```

```python
# Plot a heatmap of the correlation between different variables
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True)
plt.title('Correlation between different variables')
plt.show()
```

```python
plt.hist(df['player_id'], bins=30, edgecolor='black')
# Add labels and title
plt.xlabel('Player ID')
plt.ylabel('Goals')
plt.title('Histogram')

# Display the plot
plt.show()
```

```python
plt.figure(figsize = (10,10))

plt.subplot(2,2,1)
plt.scatter(df.index, df.assists)
plt.title('Assists')

plt.subplot(2,2,2)
plt.scatter(df.index, df.goals)
plt.title('Goals')

plt.subplot(2,2,3)
plt.scatter(df.index, df.red_cards)
plt.title('Red cards')

plt.subplot(2,2,4)
plt.scatter(df.index, df.yellow_cards)
plt.title('yellow cards')

plt.show()
```
